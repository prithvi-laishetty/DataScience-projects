## 2. Data Cleaning (5 Marks)



### 2.1 Get the column-wise Null percentage count using 'isnull()' and length of the dataframe (2 Marks)

#### Hint: When we use isnull().sum() we get sum and when we divide it by the length of the dataframe and multiply the numerator with 100 we get the percentage of missing values in each column

```python
# Get the percentage of missing values in each column .Sort the output in descending order.

Sum = hotel.isnull().sum() #detect the missing values and returns the sum of elements for each row and column.

percent = 100*(Sum/len(hotel)).sort_values(ascending = False)#logic used from the given hint and sort the values in decreasing order.

print(round(percent,2),'%n') #Round is a function returns a floating point numbers
```

### 2.2 We can see that the columns 'company' and 'agent' have 94% and 13% of values missing respectively. Since our business goals doesnot require these columns also, since these columns contain huge missing values.So,lets drop them. (2 marks)

```python
hotel = hotel.drop(labels = ['company','agent'],axis=1,errors='ignore') #It will drop the unwanted column in dataframe
```

```python
hotel.columns #Recheck the dataframe if the comoany and agent columns removed or not
```

## 3.1 Analysing the possible reasons for the customers to cancel the booking (11 Marks)

### 3.1.1 In total how many customers cancelled the booking? (2 Marks)

```python
## use the column 'is_canceled' and check the number of total cancellations

total_canceled = len(hotel[hotel['is_canceled']==1])
print("The total  number of cancelled the bookings are", total_canceled)
```

### 3.1.2 What percentage of bookings were cancelled for City Hotel and Resort Hotel? (2 marks)

```python
## Use group by and calculate the percentage of cancellations for City Hotel and Resort Hotel

#df_hotel = hotel.groupby('hotel')['is_canceled'].value_counts()
df_hotel = hotel.groupby('hotel')['is_canceled'].value_counts(normalize=True)
print(df_hotel)

#city_hotel_percent=(df_hotel['City Hotel'][1]/(df_hotel['City Hotel'].sum()))*100
city_hotel_percent = df_hotel['City Hotel'][1]*100

#resort_hotel_percent=(df_hotel['Resort Hotel'][1]/(df_hotel['Resort Hotel'].sum()))*100
resort_hotel_percent = df_hotel['Resort Hotel'][1]*100

print("\n*************************************************************************")
print("percentage of bookings cancelled for City Hotel",city_hotel_percent)
print("percentage of bookings cancelled for Resort Hotel",resort_hotel_percent)


```

### 3.1.3 Who gave more special requests? Customers who cancelled the booking or customers who didnot cancel the booking? Find the percentage for both the cases (2 Marks)

```python
# Write your code here
req_who_cancel = hotel.loc[hotel['is_canceled'] == 1, 'total_of_special_requests'].sum()
req_who_didnot_cancel = hotel.loc[hotel['is_canceled'] == 0, 'total_of_special_requests'].sum()
print(req_who_cancel)
print(req_who_didnot_cancel)
if req_who_cancel > req_who_didnot_cancel:
    print("\nTotal of special requests is more for guests who did cancel=",req_who_cancel)
else:
    print("\nTotal of special requests is more for guests who did not cancel=",req_who_didnot_cancel)
    
#percent req_who_cancel = req_who_didnot_cancel

req_canceled_percent=req_who_cancel*100/(req_who_cancel+req_who_didnot_cancel)
req_not_canceled_percent=req_who_didnot_cancel*100/(req_who_cancel+req_who_didnot_cancel)

print("% of special request by customers who cancelled the bookings",req_canceled_percent)
print("% of special request by customers who did not cancelled the bookings",req_not_canceled_percent)
```

### 3.2.2 Create a new column "total_stay_duration" in the data frame not_cancel. (2 Marks)

Total_Stay_Duration is defined as the number of nights a customer stayed in the hotel including both weekends and weekdays

```python
not_cancel['total_stay_duration'] = not_cancel['stays_in_week_nights'] + not_cancel['stays_in_weekend_nights']
not_cancel[['Customer_Id' , 'stays_in_week_nights' , 'stays_in_weekend_nights' , 'total_stay_duration']]

## total stay duration is the number of days a customer stayed in 
# the hotel including both weekends and weekdays
```

### 3.2.3 Create a new column "total_revenue" in the data frame not_cancel. (2 Marks)

Total_Revenue can be calculated by multiplying average rental revenue(adr) with total stay duration(column created in last question)

The average daily rate (adr) measures the average rental revenue earned for an occupied room per day. The operating performance of a hotel or other lodging business can be determined by using the ADR.

```python
not_cancel['total_revenue'] =  not_cancel['adr'] * not_cancel['total_stay_duration']
#not_cancel
not_cancel[['adr' , 'total_stay_duration' , 'total_revenue']]

## Adr is the average rental revenue earned. 
#If we multiply adr with total stay duration we get the total_revenue
```
